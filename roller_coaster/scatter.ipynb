```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
coasters = pd.read_csv('roller_coasters.csv')
print(coasters.head())
```

```python
def scatter(column_1, column_2, df):
    data = df[[column_1, column_2]]
    plt.scatter(data[column_1], data[column_2])
    plt.xlabel(column_1)
    plt.ylabel(column_2)
    plt.title(column_1 + ' versus ' + column_2)
scatter('length', 'speed', coasters)    
```
