```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
coasters = pd.read_csv('roller_coasters.csv')
print(coasters.head())
```

Function that build bar chart for the park needed 

```python

def park_inversions(park_name, df):
    inversions = df[df['park'] == park_name]
    inversions = inversions.sort_values('num_inversions', ascending=False)
    coaster = inversions['name']
    inv = inversions['num_inversions']
    f, ax = plt.subplots(figsize = (12, 6))
    ax.set_xticks(range(len(coaster)))
    ax.set_xticklabels(coaster, rotation=30)
    ax.bar(coaster, inv)
park_inversions('Parc Asterix', coasters)
```
