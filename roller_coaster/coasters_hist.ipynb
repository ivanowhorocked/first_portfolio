```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
coasters = pd.read_csv('roller_coasters.csv')
coasters.sort_values(by = ['speed'], ascending = False).max()
```

```python
def column_hist(column, df):
    df2 = df.dropna()
    f, ax = plt.subplots(figsize = (12, 4 ))
    plt.hist(df2[column], bins = 20, facecolor = 'seagreen')
    plt.xlabel(column)
    plt.ylabel('count')
    plt.title('Speeds of Coasters')
column_hist('speed', coasters)
```

```python
def heights_hist(df):
    heights = df[df['height'] <= 140]
    plt.hist(heights['height'], bins=20)
    plt.xlabel('height')
    plt.ylabel('count')
    plt.title('Height of Coasters')
    
heights_hist(coasters)
```
